```python
import tensorflow
import deepcut
import glob
import pandas as pd
import numpy as np
import os
import re
```

```python
text_files = glob.glob("ncpo/*.txt")

d = list()
for file in text_files:
    with open(file) as f:
        title = os.path.basename(file.split('.')[0])
        d.append(pd.DataFrame([{'title': title, 'text': f.read()}], index = [title]))      
doc = pd.concat(d)
doc.sort_values(by='title')
```

```python
txt1 = doc.text[9]
#tx = re.findall('(?<=เรื่อง)(.*)(?=สั่ง)', txt1)
tx = re.match('^.*?เรื่อง', txt1, re.DOTALL)
tx = re.findall('สั่ง.*$', txt1)
print(txt1)
#print(tx.group())
print(tx)



#regex_num = re.compile('(?<=หน้า)(.*)(?=ตอนพิเศษ)')
#s = regex_num.search(txt1)
# print(s)
```
